número $1/10$.

averiguar si el valor obtenido coincide con tu DNI.

¿Obtienes tu DNI? ¿Por qué?

```python
# Completar aquí

def multiply_and_divide(iterations):
    result =  DNI
    
    for i in range(iterations):
        result *= 1 / 10

    for i in range(iterations):
        result /= 1 / 10
    return result

DNI_2 = {multiply_and_divide(100)}
print(f" DNI_2 = {DNI_2}")

DNI == DNI_2

# Fin Completar aquí ------------------------------------
```


Consideremos la función $f(t) = j \frac{1}{1 + t}e^{ j (t + \frac{\pi}{2})}, \quad 0\leq t\leq 4\pi .$

Se pide:

1) Calcula el módulo y la fase (o argumento principal) de $f(1)$.

2) Calcula la forma binómica de $f(1)$.

3) Calcula $\frac{1}{f(1)}$ y expresa el resultado en forma binómica.

4) Calcula $f^2(1)$ y $f(1) + \frac{1}{f(1)}$ y expresa el resultado en forma binómica.

5) Haz un gráfico de $f$, es decir representa en el eje $X$ la parte real de $f$ y en el eje $Y$ la parte imaginaria.

6) Haz un gráfico de $\text{Re}f(t)$ e $\text{Im}f(t)$. 

```python
# Completar aquí

# Módulo y fase de f(1)

z = complex(0, 1 + cmath.pi / 2 )
f_1 = 1j * 0.5 * cmath.exp(z)
print('módulo = ', np.abs(f_1))
print('fase (argumento) (en radianes) = ', np.angle(f_1))
print('fase (argumento) (en grados) = ', np.rad2deg(np.angle(f_1)))

# forma binómica de f(1)

print('forma binómica de f(1) = ', f_1)

# inverso de f(1)

print('1 / f(1) = ', 1 / f_1)

# f^2(1) y 1 + 1 / f(1)

print('f^2(1) = ', f_1 ** 2)
print('f(1) + 1 / f(1) = ', f_1 + (1 / f_1))

# Gráfica de f(t)

import matplotlib.pyplot as plt

t = np.linspace(0,  4 * np.pi, 1000)
f = 1j * (1 / (1 + t)) * np.exp( 1j * (t + np.pi / 2))

x = np.real(f)
y = np.imag(f)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Gráfica de $f(t) = j 1 / (1 + t) e^{ j (t + \pi / 2)}, \quad 0\leq t\leq 4\pi $')

plt.show()

# Gráficas de las parte real e imaginaria de $f$

fig, ax = plt.subplots()
ax.plot(t, x, label='parte real de $f$')
ax.plot(t, y, label='parte imaginaria de $f$')
ax.legend()

plt.show()

# Fin Completar aquí ------------------------------------
```
