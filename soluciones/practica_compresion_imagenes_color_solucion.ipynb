# Práctica: Aplicación de la factorización SVD a la compresión de imágenes en color

En este ejercicio veremos cómo se puede comprimir una imagen a color en formato .jpg, y haciendo uso de la Descomposición en Valores Singulares (SVD).


En la carpeta **data** encontrarás el fichero **paisaje_alba.jpg**, el cual almacena un dibujo de nuestra querida artista Alba. Has de hacer los siguiente:

1) Carga dicho fichero usando el método **Image** de la librería **PIL** (recuerda que esto lo hicimos en la práctica dedicada a tensores) y guárdalo en una variable que llamaremos $M$.

2) Transforma dicho fichero en un array de numpy. Llámalo también M. Calcula la forma de dicho array, su tamaño, y el tipo de datos que contiene.

3) Transforma de nuevo el array M a una imagen del módulo PIL (revisa la práctica de tensores). Llama **paisaje_alba** a la imagen resultante, y muéstrala por pantalla.



```python
# Completar aquí

import numpy as np

from PIL import Image

M = Image.open("../data/paisaje_alba.jpg")
M = np.array(M)

#M = np.load("../data/tensor_nasa.npy")
print(f"forma del array =  {M.shape}")
print(f"tamaño del array =  {M.size}")
print(f"tipo de datos del tensor = {M.dtype}")

paisaje_alba = Image.fromarray(M)

paisaje_alba.show()

M = M.astype(np.float64)
print(f"tipo de datos convertidos = {M.dtype}")

# Fin Completar aquí ------------------------------------
```

El objetivo que nos proponemos es comprimir dicha imagen usando la factorización SVD. Para ello, la estrategia que vamos a seguir es: (a) vamos a extraer del tensor $M$ las tres matrices correspondientes a cada uno de los colores **R, G, B**, (b) haremos la factorización SVD de cada una de estas matrices, y (c) recompondremos la imagen en color.

Vayamos por partes.

1) Utiliza la técnica de slicing para extraer del tensor $M$ las tres matrices que lo componen. Denota a dichas matrices por $M_{\text{r}}$, $M_{\text{g}}$, y $M_{\text{b}}$, respectivamente, e imprime sus formas.



```python
# Completar aquí

M_r = M[:, :, 0]
print(f"forma de M_r = {M_r.shape}")
#imagen_red = Image.fromarray(M_r)
#imagen_red.show()
M_g = M[:, :, 1]
print(f"forma de M_g = {M_g.shape}")
#imagen_green = Image.fromarray(M_g)
#imagen_green.show()
M_b = M[:, :, 2]
print(f"forma de M_b = {M_b.shape}")
#imagen_blue = Image.fromarray(M_b)
#imagen_blue.show()

# Fin Completar aquí ------------------------------------
```

Vamos a utilizar la descomposición en valores singulares (SVD) para comprimir las tres matrices anteriores.

Así pues, calcula la factorización SVD de las matrices $M_r$, $M_g$, $M_b$. Denota adecuadamente a las matrices $U$, $\Sigma$ y $V^T$ resultantes de estas factorizaciones. 

Calcula también el tipo de datos que almacenan las matrices resultantes de la factorización SVD. Es suficiente con que lo hagas con  matriz $U$ de $M_r$.



```python
# Completar aquí

from scipy.linalg import svd

# Matriz M_r
U_r, s_r, V_r_T = svd(M_r)
Sigma_r = np.zeros((M_r.shape[0], M_r.shape[1]))
Sigma_r[:M_r.shape[0], :M_r.shape[0]] = np.diag(s_r)

print(f"tipo de datos en U_r = {U_r.dtype}")

# Matriz M_g
U_g, s_g, V_g_T = svd(M_g)
Sigma_g = np.zeros((M_g.shape[0], M_g.shape[1]))
Sigma_g[:M_g.shape[0], :M_g.shape[0]] = np.diag(s_g)

# Matriz M_b
U_b, s_b, V_b_T = svd(M_b)
Sigma_b = np.zeros((M_b.shape[0], M_b.shape[1]))
Sigma_b[:M_b.shape[0], :M_b.shape[0]] = np.diag(s_b)

# Fin Completar aquí ------------------------------------
```

Usando la técnica de **slicing** vamos a quedarnos con los valores singulares mayores de cada una de las matrices $\Sigma$ anteriores. Para ello:

1) Introduce una variable, llamada **split**, con valor **split** $ = 150$.

Utilizando la técnica de slicing: 

2) Extrae de las matrices $U$ las primeras **split** columnas. 

3) Genera nuevas matrices $\Sigma_{\text{split}}$ extrayendo las primeras **split** filas y las primeras **split** columnas de las $\Sigma$ anteriores. 

4) Extrae de las matrices $V^T$ anteriores las primeras **split** filas.

5) Imprime las formas de las matrices anteriores.



```python
# Completar aquí

split = 150
# Matriz r
U_r_split = U_r[:, :split]
print(f"shape de U_r_split = {U_r_split.shape}")
Sigma_r_split = Sigma_r[:split, :split]
print(f"shape de Sigma_r_split = {Sigma_r_split.shape}")
V_r_split = V_r_T[:split, :]
print(f"shape de V_r_split = {V_r_split.shape}")

# Matriz r
U_g_split = U_g[:, :split]
print(f"shape de U_g_split = {U_g_split.shape}")
Sigma_g_split = Sigma_g[:split, :split]
print(f"shape de Sigma_g_split = {Sigma_g_split.shape}")
V_g_split = V_g_T[:split, :]
print(f"shape de V_g_split = {V_g_split.shape}")

# Matriz b
U_b_split = U_b[:, :split]
print(f"shape de U_b_split = {U_b_split.shape}")
Sigma_b_split = Sigma_b[:split, :split]
print(f"shape de Sigma_b_split = {Sigma_b_split.shape}")
V_b_split = V_b_T[:split, :]
print(f"shape de V_b_split = {V_b_split.shape}")


# Fin Completar aquí ------------------------------------
```

Genera las matrices de rango menor o igual a **split** que mejor aproximan a las matrices originales $M_r$, $M_g$ y $M_b$ anteriores.


```python
# Completar aquí

M_r_split = U_r_split @ Sigma_r_split @ V_r_split

M_g_split = U_g_split @ Sigma_g_split @ V_g_split

M_b_split = U_b_split @ Sigma_b_split @ V_b_split

# Fin Completar aquí ------------------------------------
```

1) A parir de las tres matrices anteriores, construye un tensor de orden $3$ formado por dichas matrices, calcula su forma y su tamaño.

2) Transforma el tensor anterior en una imagen del módulo PIL que has de llamar **image_compressed**.



```python
# Completar aquí

image_array = np.stack((M_r_split, M_g_split, M_b_split), axis=-1)
image_array_uf8 = np.clip(image_array, 0, 255).astype(np.uint8)
print(f"forma de image_array = {image_array.shape}")
image_compressed = Image.fromarray(image_array_uf8)
image_compressed.show()

# Fin Completar aquí ------------------------------------
```

Observarás que muy posiblemente te muestre un mensaje de error. 
El motivo es que la imagen original trabaja con un tipo de datos que la factorización SVD ha transformado. Para solucionar este problema, una vez has transformado la imagen original en un tensor, cambia el tipo de datos del tensor original con **M = M.astype(np.float64)**

Finalmente, cuando hayas reconstruido el tensor que contiene la imagen comprimida, devuelve los datos al tipo original con **image_array_uf8 = np.clip(image_array, 0, 255).astype(np.uint8)**



Si grabas ahora la imagen comprimida (puedes hacerlo con **image_compressed.save("../data/paisaje_alba_comprimido.jpg", "JPEG")** ) observarás que la imagen comprimida ocupa **100KB** mientras que la imagen sin comprimir original ocupa **242KB**.

```python
image_compressed.save("../data/paisaje_alba_comprimido.jpg", "JPEG")
```
